```python
#Congélateur
import numpy as np
import pandas as pd

puissance_nominal = 120  # Puissance nominale en Watts
cos_fi_nominal = 0.8  # Facteur de puissance nominal

# Générer les valeurs de tension et de courant
t = np.linspace(0, 1, num=65536)* 3600  # Temps de 0 à 1 heure en secondes
u = np.random.normal(loc=220, scale=10, size=len(t))  # Tension aléatoire
i = puissance_nominal / (u * cos_fi_nominal)  # Courant RMS nominal à partir de la puissance et du facteur de puissance nominal

cos_fi = np.cos(np.arccos(cos_fi_nominal) + np.pi * (np.random.rand() - 0.5))  # Cos(fi) aléatoire à partir du facteur de puissance nominal
p = u * i * cos_fi * np.random.randn(len(t)) * np.sqrt(len(t) / 500)  # Puissance aléatoire pour chaque échantillon de données

# Sauvegarder les données de la puissance dans un fichier CSV
p = pd.DataFrame(p)
p.to_csv('Puissance_congelateur.csv', header=False, index=False)

# Charger le fichier CSV contenant la puissance
puissance = pd.read_csv('Puissance_congelateur.csv', header=None)

# Afficher les valeurs de puissance
print(puissance.values)

# Charger le fichier CSV
data = pd.read_csv('Puissance_congelateur.csv')

# Récupérer les valeurs de puissance sous forme de tableau numpy
p = data.to_numpy()

# Afficher les valeurs de puissance sans crochets
for val in p:
    print(val[0], end=' ')
```
