```python
#Ordinateur en charge
import csv
import numpy as np
import pandas as pd

amplitude = 220  # Amplitude de la tension en Volts
frequence = 50  # Frquence en Hz
phase = 0  # Phase en degrs

# Gnrer les valeurs de tension et de courant
t = np.linspace(0, 1, num=65536)* 3600  # Temps de 0  1 heure en secondes
u = amplitude * np.sin(2 * np.pi * frequence * t)  # Tension sinusoidale
i = amplitude * np.sin(2 * np.pi * frequence * t + np.deg2rad(phase))  # Courant sinusoidal avec la phase dfinie

cos_fi = np.cos(np.deg2rad(phase))  # Calcul du cos(fi)  partir de la phase dfinie
n_samples = 65536
p = u * i * cos_fi * np.random.randn(n_samples) * np.sqrt(len(t) / 500)  # Puissance alatoire pour n_samples chantillons de donnes

# Sauvegarder les donnes de la puissance dans un fichier CSV
p = pd.DataFrame(np.reshape(p, (-1, 256)))
p.to_csv('Puissance.csv', header=False, index=False)

# Charger le fichier CSV contenant la puissance
puissance = pd.read_csv('Puissance.csv', header=None)

# Afficher les valeurs de puissance
print(puissance.values)
# Charger le fichier CSV
data = pd.read_csv('Puissance.csv')

# Rcuprer les valeurs de puissance sous forme de tableau numpy
p = data.to_numpy()

# Afficher les valeurs de puissance sans crochets
for row in p:
    for val in row:
        print(val, end=' ')
```
