```python
#Bouilloire
import numpy as np

puissance_nominal = 2000  # Puissance nominale en Watts
u_nominal = 220  # Tension nominale en Volts
cos_fi_nominal = 1  # Facteur de puissance nominal (ici, on suppose que la bouilloire est purement resistive)

# Générer les valeurs de tension et de courant
t = np.linspace(0, 1, num=65536) * 3600  # Temps de 0 à 1 heure en secondes
u = u_nominal * np.sin(2 * np.pi * 50 * t)  # Tension sinusoidale
i = puissance_nominal / u  # Courant RMS nominal à partir de la puissance et de la tension nominale
p = np.abs(u * i * cos_fi_nominal * np.random.randn(len(t)) * np.sqrt(len(t) / 500))  # Puissance aléatoire pour chaque échantillon de données

# Sauvegarder les données de la puissance dans un fichier CSV
np.savetxt('puissance_bouilloire.csv', p, delimiter=',')

# Charger le fichier CSV contenant la puissance
puissance = np.loadtxt('puissance_bouilloire.csv', delimiter=',')

# Afficher les valeurs de puissance
print(puissance)

# Afficher les valeurs de puissance sans crochets
for val in puissance:
    print(val, end=' ')

```
